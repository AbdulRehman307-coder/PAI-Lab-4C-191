### Water Jug

```python
jugA_capacity = 4
jugB_capacity = 3
goal = 2

visited = set()

def dfs(state, path):
    a, b = state

    # Goal check
    if a == goal or b == goal:
        print(" Goal Reached!")
        for step in path:
            print(step)
        return True

    if state in visited:
        return False

    visited.add(state)

    # Possible actions
    actions = []

    # 1. Fill Jug A
    actions.append(("Fill Jug A", (jugA_capacity, b)))

    # 2. Fill Jug B
    actions.append(("Fill Jug B", (a, jugB_capacity)))

    # 3. Empty Jug A
    actions.append(("Empty Jug A", (0, b)))

    # 4. Empty Jug B
    actions.append(("Empty Jug B", (a, 0)))

    # 5. Pour A -> B
    pour = min(a, jugB_capacity - b)
    actions.append(("Pour Jug A → Jug B", (a - pour, b + pour)))

    # 6. Pour B -> A
    pour = min(b, jugA_capacity - a)
    actions.append(("Pour Jug B → Jug A", (a + pour, b - pour)))

    for rule, new_state in actions:
        if new_state not in visited:
            if dfs(new_state, path + [f"{rule}: {state} → {new_state}"]):
                return True

    return False


# Start DFS
start_state = (0, 0)
dfs(start_state, ["Start: (0, 0)"])
```
